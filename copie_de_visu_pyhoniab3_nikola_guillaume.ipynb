#Heatmap

```python
# Calcul des corrélations
corr = df[['Kills Team' , 'Turrets Team', 'Dragon Team', 'Baron Team',	'Level', 'Kills','Deaths','Assists','CS','Golds','Total damage to Champion','Physical Damage','Magic Damage','True Damage','Items Purchased']].corr()

corr_forte= corr[abs(corr) > 0.7]


# Heatmap

sns.heatmap(corr_forte, annot=True, cmap='coolwarm')
plt.show()

#Kills Team , Turrets Team, Dragon Team, Baron Team,	Level, Kills,	Deaths,	Assists,	CS, Golds, Total damage to Champion,	Physical Damage,	Magic Damage,	True Damage,Items Purchased
```

Dans cette heatmap on à sortie les variables ayant une corélation forte on peut voir que celles avec la corélation la plus élevé sont les CS et les Gold, les CS et le Level, Level et les Gold, et les Gols et le total dammage to champion . Pour une analyse plus poussé de la base de donné ces champs pourraient donc être supprimer pour laisser seuelement le Gold.
